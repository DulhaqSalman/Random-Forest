```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc=StandardScaler()
# x=sc.fit_transform(x_train)
# x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
# x=sc.fit_transform(x_train)
x_t=sc.fit(x_train)
x_t
```

```python
x_train=x_t.transform(x_train)
x_train
```

```python
x_test=x_t.transform(x_test)
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=11,criterion="entropy",random_state=0)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
classifier.score(x_train,y_train)
```

```python
classifier.score(x_train,y_train)
```

```python
feature_col=data
feature_col=feature_col.drop('spectype',axis=1)
```

```python
from sklearn import tree
plt.figure(dpi=300)
tree.plot_tree(classifier.estimators_[1],feature_names=feature_col.columns,
              class_names=["setosa","versicolor","virginica"],filled=True,rounded=True,fontsize=4)
plt.show()
```
